# Matched filter SNR

We give more details about the matched filter signal-to-noise ratio (SNR) statistics used to detect 
gravitational wave signals.

The matched filter SNR statistic is defined as
$$\hat{\rho} = \frac{\hat{s}}{N} =  \frac{(h | d)}{\sqrt{(h | h)}}$$
where $h$ is the signal template (or model) and $d$ is the observed data.
We noted $(a|b)$ is the noise-weighted inner product:
$$ ( a | b ) = 4 \mathrm{Re} \int_{0}^{+\infty} df \; \frac{\tilde{a}^{*}(f)\tilde{b}(f)}{S_n(f)}$$
where $\tilde{a}(f)$ and $\tilde{b}(f)$ are the Fourier transforms of $a$ and $b$ and $S_{n}(f)$ 
is the noise one-sided power spectral density (PSD).

The signal template $h$ usually depends on parameters $\theta$ that need to be adjusted to maximize the matched
filter SNR. In the matched filtering technique presented here, we assume that the data $d$ is the 
sum of a signal $h(\theta_{\mathrm{true}})$ and a stationary noise $n$ described by a PSD $S_{n}(f)$ so that
$$ d = h(\theta_{\mathrm{true}}) + n $$
where $h(\theta_{\mathrm{true}})$ is the true value of the parameters describing the signal.

We also define the optimal SNR, which is obtained by evalutating the matched filter SNR in the 
absence of noise and using the right template, so that the model matches exactly what the data
contains: $n=0$, $h=h(\theta_{\mathrm{true}})$:
$$\rho_{\mathrm{opt}} = \sqrt{(h(\theta_{\mathrm{true}}) | h(\theta_{\mathrm{true}}))}$$

If the noise $n$ is stationary, Gaussian, and is described by the PSD $S_{n}(f)$, we have:
* In the presence of noise only: $\hat{\rho} \sim \mathcal{N}(0,1)$
* When there is a signal+noise: $\hat{\rho} \sim \mathcal{N}(\rho_{\rm opt},1)$


As detector data are discretized in time, we can implement the noise-weighted inner product as
$$ ( a | b ) = 4 \Delta f \mathrm{Re} \sum_{k=k_{\min}}^{k_{\max}} \frac{\tilde{a}_{k}^{*} \tilde{b}_{k}}{S_{n}(k \Delta f)} $$
where $\Delta f$ is the frequency resolution of the data, related to the observation time as
$$\Delta f = \frac{1}{T_{\mathrm{obs}}}$$
The minimum and maximum frequency bins $k_{\min}$ and $k_{\max}$ can be restricted to the frequency
band of interest, and do not nead to span the full domain from 0 to the Nyquist frequency.

Note that the inner product can be re-written in terms of whitened, possibly band-passed, data,
by defining:
$$ \tilde{a}_{\rm wbp} (f) = w(f) \times \sqrt{\frac{2 \Delta f}{S_n(f)}} \tilde{a}(f)$$
where $w(f)$ is the frequency response of the band-pass filter such that $|w(f)|^2$ is close to 1 in the region 
of interest and close to 0 for excluded frequencies.
We can then write
$$ ( a | b ) = \sum_{k=k_{\min}}^{k_{\max}} \tilde{a}_{\rm wbp}^{*}(k \Delta f) \tilde{b}_{\rm wbp}(k \Delta f) $$
